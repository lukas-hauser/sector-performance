```python
# Automobiles & Parts SXAP
SXAP = data.DataReader('^SXAP', 'yahoo', start, end)

# Banks
SX7P = data.DataReader('^SX7P', 'yahoo', start, end)

# Basic Resources SXPP
SXPP = data.DataReader('^SXPP', 'yahoo', start, end)

# Chemicals SX4P
SX4P = data.DataReader('^SX4P', 'yahoo', start, end)

# Construction & Materials SXOP
SXOP = data.DataReader('^SXOP', 'yahoo', start, end)

# Food & Beverage SX3P
SX3P = data.DataReader('^SX3P', 'yahoo', start, end)

# Health Care SXDP
SXDP = data.DataReader('^SXDP', 'yahoo', start, end)

# Industrial Goods & Services SXNP
SXNP = data.DataReader('^SXNP', 'yahoo', start, end)

# Insurance SXIP
SXIP = data.DataReader('^SXIP', 'yahoo', start, end)

# Media SXMP
SXMP = data.DataReader('^SXMP', 'yahoo', start, end)

# Oil & Gas SXEP
SXEP = data.DataReader('^SXEP', 'yahoo', start, end)

# Personal & Household Goods SXQP
SXQP = data.DataReader('^SXQP', 'yahoo', start, end)

# Retail SXRP
SXRP = data.DataReader('^SXRP', 'yahoo', start, end)

# Technology SX8P
SX8P = data.DataReader('^SX8P', 'yahoo', start, end)

# Telecommunications SXKP
SXKP = data.DataReader('^SXKP', 'yahoo', start, end)

# Travel & Leisure SXTP
SXTP = data.DataReader('^SXTP', 'yahoo', start, end)

# Utilities SX6P
SX6P = data.DataReader('^SX6P', 'yahoo', start, end)
```

```python
sectors_symbols = ['SXAP','SX7P','SXPP','SX4P','SXOP','SX3P','SXDP','SXNP','SXIP','SXMP','SXEP','SXQP','SXRP','SX8P','SXKP','SXTP','SX6P']
```

```python
sectors = [      'Automobiles & Parts',
                 'Banks',
                 'Basic Resources',
                 'Chemicals',
                 'Construction & Materials',
                 'Food & Beverage',
                 'Health Care',
                 'Industrial Goods & Services',
                 'Insurance',
                 'Media',
                 'Oil & Gas',
                 'Personal & Household Goods',
                 'Retail',
                 'Technology',
                 'Telecommunications',
                 'Travel & Leisure',
                 'Utilities']
```

```python
sector_indices = pd.concat([SXAP,SX7P,SXPP,SX4P,SXOP,SX3P,SXDP,SXNP,SXIP,SXMP,SXEP,SXQP,SXRP,SX8P,SXKP,SXTP,SX6P],axis=1,keys=sectors)
#sector_indices.tail()
```

```python
sector_indices.columns.names = ['Sector Symbol','Sector Info']
```

```python
sector_indices.head()
```

```python
returns = pd.DataFrame()
```

```python
for sector in sectors:
    returns[sector] = sector_indices[sector]['Close'].pct_change()
returns.head()
```

```python
cum_returns = (1 + returns).cumprod()
```
